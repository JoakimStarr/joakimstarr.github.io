```python
import pandas as pd
```

```python
# 读取D:\paper\企业财务数据下面的每一个文件，并将文件名和路径保留为数组
import os
path = 'D:/paper/企业财务数据'
files = []
for file in os.listdir(path):
    # 保留文件名和路径
    files.append(os.path.join(path, file))

files
```

```python
# 将dta文件读取并转为excel文件
for file in files:
    # 读取dta文件
    df = pd.read_stata(file)
    # 将文件名中的.dta替换为.xlsx
    new_file = file.replace('.dta', '.xlsx')
    # 将数据保存为excel文件
    df.to_excel(new_file, index=False)
```
